```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(title, fn):
  x = np.arange(-5, 5, 0.05)
  y = [fn(xi) for xi in x]
  plt.plot(x, y)
  plt.title(title)
  plt.show()
```

```python
def relu(x):
  return max(0.0, x)

plot('ReLu', relu)
```

```python
def sigmoid(x, a=1):
  return np.divide(
      1,
      (
          1 + np.exp(
              -(np.multiply(a, x))
          )
      )
  )

plot('Sigmoid', sigmoid)
```

```python
def tanh(x, a=1, b=1):
  return np.tanh(x)

plot('Hyperbolic tangent', tanh)
```

```python
def linear(x):
  return x

plot('Linear', linear)
```
